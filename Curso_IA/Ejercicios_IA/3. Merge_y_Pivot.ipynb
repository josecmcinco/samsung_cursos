```python
import pandas as pd
import numpy as np

# Load Data
userHeader = ['user_id', 'gender', 'age', 'ocupation', 'zip']
users = pd.read_table('https://raw.githubusercontent.com/jarroba/Example_Pandas/master/dataSet/users.txt', engine='python', sep='::', header=None, names=userHeader, encoding="ISO-8859-1")

movieHeader = ['movie_id', 'title', 'genders']
movies = pd.read_table('https://raw.githubusercontent.com/jarroba/Example_Pandas/master/dataSet/movies.txt', engine='python', sep='::', header=None, names=movieHeader, encoding="ISO-8859-1")

ratingHeader = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('https://raw.githubusercontent.com/jarroba/Example_Pandas/master/dataSet/ratings.txt', engine='python', sep='::', header=None, names=ratingHeader, encoding="ISO-8859-1")
```

Combina los 3 dataframes

```python
df_sucio = pd.merge(users, ratings, on="user_id") # Unión de usuario con rating

df = pd.merge(df_sucio, movies, on="movie_id") # DataFrame final

df.head()

```

Mostrar la edad media y rating medio para cada película

```python
pd.pivot_table(df, index="movie_id",values = ["rating", "age"] )
```

Mostrar para cada película la suma de ratings, el número de ratings y la media de rating

```python
pd.pivot_table(df, index = ["movie_id"], values = ["rating"], aggfunc=["sum", np.size, np.mean] )
```

Muestra la media de rating para cada película, separado por género

```python
pd.pivot_table(df, index = ["movie_id"], values = ["rating"], aggfunc=["sum", np.size, np.mean])
```
