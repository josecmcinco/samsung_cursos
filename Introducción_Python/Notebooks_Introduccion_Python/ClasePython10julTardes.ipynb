## Atributos/métodos con doble guión bajo

```python
import math
class Circulo:
    def __init__(self, radio):
        self.__radio = radio
    def area(self):
        return math.pi * self.__radio ** 2
```

```python
c = Circulo(12)
```

```python
c.__dict__ # podemos observar el 'name mangling'
```

```python
c.area()
```

```python
c._Circulo__radio = 1
```

```python
c.area()
```

### Herencia

```python
class A:
    def metodo(self):
        print("soy un método en A")
```

```python
a = A()
a.metodo()
```

```python
class B(A): # B hereda de A (A es padre de B)
    def metodo(self):
        print("soy un método de B")
```

```python
b = B()
b.metodo()
```

```python
isinstance(b,B)
```

```python
isinstance(a,B)
```

```python
isinstance(b,A)
```

```python
isinstance(True,int)
```

```python
issubclass(bool,int)
```

```python
issubclass(B,A)
```

```python
class C(B):
    pass
```

```python
c = C()
```

```python
c.metodo()
```

```python
issubclass(C,A)
```
