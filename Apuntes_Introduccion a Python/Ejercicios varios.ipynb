```python
scores = [100, 90, 95, 90, 80, 70, 0, 80, 90, 90, 0, 90, 90, 0, 90, 100, 75, 20, 30, 50, 90]

iterScores = iter(scores)

totalStudents = len(scores)//3

respuesta = []

for i in range(totalStudents):
    notaIng = next(iterScores)
    notaMat = next(iterScores)
    notaCien = next(iterScores)
    if (notaIng and notaMat and notaCien):
        respuesta.append([notaIng, notaMat, notaCien])
        
print("Estudiantes totales. ", totalStudents)
print("Estudiantes validos. ", len(respuesta))
print(respuesta)
```

```python
doc = open("valores.txt", "r", encoding= "Utf-8")

lines = doc.readlines()

respuesta = 0.0

for linea in lines:
    palabras = linea.split()
    for palabra in palabras:
        try:
            palabra = float(palabra)
            respuesta += palabra
        except:
            pass

doc.close()

print(respuesta)
```

```python
from functools import reduce

doc = open("el_quijote.txt", "r", encoding= "Utf-8")

dict1 = {}

lineas = doc.readlines()

for linea in lineas:
    linea.replace(",", "")
    linea.replace(".", "")

    linea.strip()

    palabraMaslarga = ""

    palabras = linea.split()

    palabraMaslarga = reduce(lambda acc, x: x if len(x) > len(acc) or (len(x) == len(acc) and x[0] > acc[0]) else acc, palabras, "")

    if palabraMaslarga not in dict1:
        dict1[palabraMaslarga] = 1

    else:
        dict1[palabraMaslarga] += 1



doc.close()

dict1
```
